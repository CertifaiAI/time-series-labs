```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
airpassengers = pd.read_csv('../datasets/decomposition/AirPassengers.csv')

airpassengers_series = pd.Series(airpassengers['#Passengers'].values, 
                            index = pd.date_range('1949-01', periods = len(airpassengers), freq='M'))

airpassengers_series = np.log(airpassengers_series)
```

```python
airpassengers_series.plot()
```

```python
from statsmodels.tsa.seasonal import STL

nybirths_stl = STL(airpassengers_series).fit()
nybirths_stl.plot()
plt.show()
```

```python
data = nybirths_stl.seasonal
```

```python
data.plot()
```

```python
df = pd.DataFrame({
    'y': data,
    't': data.shift(1),
    't-1': data.shift(2),
    't-2': data.shift(3),
    't-3': data.shift(4)
})
```

```python
df.head()
```

```python
df2 = df['1949-05-31':]
```

```python
df2.head()
```

```python
df2.tail()
```

```python
train = df2[:'1959-12-31']
train
```

```python
test = df2['1960-01':]
test
```

```python
train_X, train_y = train.iloc[:,1:].values, train['y']
test_X, test_y = test.iloc[:,1:].values, test['y']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(train_X)
train_X_scaled = scaler.transform(train_X)
test_X_scaled = scaler.transform(test_X)
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
```

```python
lr = LinearRegression().fit(train_X_scaled, train_y)
```

```python
lr.coef_
```

```python
pred = lr.predict(test_X_scaled)
```

```python
pred_series = pd.Series(pred, index=test.index)
```

```python
plt.plot(train['y'])
plt.plot(test['y'])
plt.plot(pred_series)
```
